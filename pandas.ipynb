```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sales.csv", encoding='latin1')


df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])


total_sales = df['SALES'].sum()
print("üìä Total Sales: $", round(total_sales, 2))


sales_by_year = df.groupby('YEAR_ID')['SALES'].sum()
print("\nüìÖ Sales by Year:")
print(sales_by_year)


sales_by_country = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)
print("\nüåç Sales by Country:")
print(sales_by_country)


sales_by_product = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
print("\nüì¶ Sales by Product Line:")
print(sales_by_product)


sales_by_dealsize = df.groupby('DEALSIZE')['SALES'].sum()
print("\nüìÅ Sales by Deal Size:")
print(sales_by_dealsize)
```

```python
# Sales by Year
sales_by_year.plot(kind='bar', title='Sales by Year', color='skyblue')
plt.ylabel("Total Sales")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

```python
# Sales by Product Line
sales_by_product.plot(kind='barh', title='Sales by Product Line', color='orange')
plt.xlabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#  Sales by Country (Top 10)
sales_by_country.head(10).plot(kind='bar', title='Top 10 Countries by Sales', color='green')
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Sales by Deal Size
sales_by_dealsize.plot(kind='pie', title='Sales by Deal Size', autopct='%1.1f%%', startangle=90)
plt.ylabel("")
plt.show()
```
